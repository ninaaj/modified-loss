```python
import tensorflow as tf

from tensorflow import keras
```

custom accuracies and true and false positives and negatives

these work with labels that have category values attached

```python
@keras.saving.register_keras_serializable(package='custom_metrics')
class CTP(tf.keras.metrics.Metric):

    def __init__(self, name='custom_true_positives', dtype=None):
        super().__init__(
          name=name, dtype=dtype
        )
        self.tp = tf.keras.metrics.TruePositives()


    def update_state(self, y_true, y_pred, threshold=0.5, sample_weight=None):

        self.tp.update_state(tf.gather(y_true,[0],axis=1), y_pred)

    def result(self):

        return self.tp.result()

    def reset_state(self):

        self.tp.reset_state()

    def get_config(self):

        return  {'name': 'custom_true_positives', 'dtype': 'float32'}

@keras.saving.register_keras_serializable(package='custom_metrics')
class CTN(tf.keras.metrics.Metric):

    def __init__(self, name='custom_true_negatives', dtype=None):
        super().__init__(
          name=name, dtype=dtype
        )
        self.tn = tf.keras.metrics.TrueNegatives()

    def update_state(self, y_true, y_pred, threshold=0.5, sample_weight=None):

        self.tn.update_state(tf.gather(y_true,[0],axis=1), y_pred)

    def result(self):

        return self.tn.result()

    def reset_state(self):

        self.tn.reset_state()

    def get_config(self):

        return {'name': 'custom_true_negatives', 'dtype': 'float32'}

@keras.saving.register_keras_serializable(package='custom_metrics')
class CFP(tf.keras.metrics.Metric):

    def __init__(self, name='custom_false_positives', dtype=None):
        super().__init__(
          name=name, dtype=dtype
        )
        self.fp = tf.keras.metrics.FalsePositives()


    def update_state(self, y_true, y_pred, threshold=0.5, sample_weight=None):

        self.fp.update_state(tf.gather(y_true,[0],axis=1), y_pred)

    def result(self):

        return self.fp.result()

    def reset_state(self):

        self.fp.reset_state()

    def get_config(self):

        return {'name': 'custom_false_positives', 'dtype': 'float32'}

@keras.saving.register_keras_serializable(package='custom_metrics')
class CFN(tf.keras.metrics.Metric):

    def __init__(self, name='custom_false_negatives', dtype=None):
        super().__init__(
          name=name, dtype=dtype
        )
        self.fn = tf.keras.metrics.FalseNegatives()

    def update_state(self, y_true, y_pred, threshold=0.5, sample_weight=None):

        self.fn.update_state(tf.gather(y_true,[0],axis=1), y_pred)

    def result(self):

        return self.fn.result()

    def reset_state(self):

        self.fn.reset_state()

    def get_config(self):

        return {'name': 'custom_false_negatives', 'dtype': 'float32'}

@keras.saving.register_keras_serializable(package='custom_metrics')
class CBA(tf.keras.metrics.Metric):

    def __init__(self, name='custom_binary_accuracy', dtype=None):
        super().__init__(
          name=name, dtype=dtype
        )
        self.accuracy = tf.keras.metrics.BinaryAccuracy()

    def update_state(self, y_true, y_pred, threshold=0.5, sample_weight=None):

        self.accuracy.update_state(tf.gather(y_true,[0],axis=1), y_pred)

    def result(self):

        return self.accuracy.result()

    def reset_state(self):

        self.accuracy.reset_state()

    def get_config(self):

        return {'name': 'custom_binary_accuracy', 'dtype': 'float32'}

@keras.saving.register_keras_serializable(package='custom_metrics')
class CCA(tf.keras.metrics.Metric):

    def __init__(self, name='custom_categorical_accuracy', dtype=None, number_of_classes=None):
        super().__init__(
          name=name, dtype=dtype
        )

        self.accuracy = tf.keras.metrics.CategoricalAccuracy()

        self.number_of_classes = number_of_classes

    def update_state(self, y_true, y_pred, threshold=0.5, sample_weight=None):

        self.accuracy.update_state(tf.gather(y_true,range(self.number_of_classes),axis=1), y_pred)

    def result(self):

        return self.accuracy.result()

    def reset_state(self):

        self.accuracy.reset_state()

    def get_config(self):

        return {'name': 'custom_categorical_accuracy', 'dtype': 'float32','number_of_classes': self.number_of_classes}
```
